```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data, labels = make_blobs(n_samples=100, n_features=2, centers=3, random_state=1)
```

```python
print('blaze::DynamicMatrix<float> data{',)
for i in range(data.shape[0]):
    print('  { ', end='')
    for j in range(data.shape[1]):
        print(data[i,j], end='')
        print('F, ', end='')
    print('},')
    
print('\n}')
```

```python
centroidsKMeans = np.array([
      [ -10.0494,     -3.85954],
      [ -1.47108,      4.33722],
      [ -6.58197,     -8.17239],
])

centroidsLocalSearch = np.array([
    (     -1.34052,      4.15712 ),
    (     -6.53542,     -8.01553 ),
    (     -10.3416,     -3.90975 ),
])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(data[:,0], data[:,1], c=labels)
ax.scatter(centroidsKMeans[:,0], centroidsKMeans[:,1], marker='x', color='red', s=100, label="k-Means centers")
ax.scatter(centroidsLocalSearch[:,0], centroidsLocalSearch[:,1], marker='x', color='#00ffff', s=100, label='Local Search centers')
ax.legend();
```
