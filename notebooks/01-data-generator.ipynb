```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data, labels = make_blobs(n_samples=100, n_features=2, centers=3, random_state=1)
```

```python
print('blaze::DynamicMatrix<float> data{',)
for i in range(data.shape[0]):
    print('  { ', end='')
    for j in range(data.shape[1]):
        print(data[i,j], end='')
        print('F, ', end='')
    print('},')
    
print('\n}')
```

```python
centroids = np.array([
      [ -10.0494,     -3.85954],
      [ -1.47108,      4.33722],
      [ -6.58197,     -8.17239],
])
```
